```python
#q1 - In this assignment students have to transform iris data into 3 dimensions
# and plot a 3d chart with transformed dimensions and colour each data
# point with specific class.


import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets

#Loading dataset
df=datasets.load_iris()
#df
```

```python
x = decomposition.PCA(n_components=3).fit_transform(df.data)
y = df.target

fig=plt.figure(figsize=(16,10))
ax=plt.axes(projection='3d')

ax.scatter(x[:, 0], x[:, 1], x[:, 2], c=y, cmap=plt.cm.rainbow_r)

ax.set_title("First 3 PCA Directions")
ax.set_xlabel("1D")
ax.set_ylabel("2D")
ax.set_zlabel("3D")

plt.tight_layout()
```
